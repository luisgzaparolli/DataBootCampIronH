#### Find the number of titles that have changed in the United States Code since its last release point.

````python
# This is the url you will scrape in this exercise
url = 'http://uscode.house.gov/download/download.shtml'
````

````python
# your code here
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html)
uschanged = []
for item in soup.find_all('div', attrs={'class':'uscitem'}):
    for titles in item.find_all('div', attrs={'class': 'usctitlechanged'}):
        for rawtitle in titles:
            try:
                rawtitle = rawtitle.strip()
                uschanged.append(rawtitle)
            except:
                pass
for item in range(len(uschanged)):
    uschanged.remove('')
uschanged
````

#### Count the number of tweets by a given Twitter account.
Ask the user for the handle (@handle) of a twitter account. You will need to include a ***try/except block*** for account names not found. 
<br>***Hint:*** the program should count the number of tweets for any provided account.

````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url
url = 'https://twitter.com/'
````

````python
# your code here
def count_tweets():
    user = input('Enter a twitter name: ')
    url = 'https://twitter.com/'
    url = url + user
    response = requests.get(url)
    if response.status_code == 200:
        html = response.content
        soup = BeautifulSoup(html)
        followers_nav = soup.find('li', {'class':'ProfileNav-item ProfileNav-item--tweets is-active'})
        follower = int(followers_nav.find('a').find('span', {'class':'ProfileNav-value'})['data-count'])
        print(f'{user} have {follower} followers.')
    else:
        print('Account name not found!')
````

````python
count_tweets()
````

#### Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
# your code here
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html)
table = soup.find_all('table', attrs = {'class': 'chart'})[0]
headers = ['movie_name', 'initial_release', 'director_name', 'stars']
title = [title.find('a').text for title in table.find_all('td', attrs = {'class': 'titleColumn'})]
initial_release = [release.text.strip().replace('(','').replace(')','') for release in table.find_all('span', attrs = {'class': 'secondaryInfo'})]
directoractor_list = [director.find('a')['title'].split(',') for director in table.find_all('td', attrs = {'class': 'titleColumn'})]
director_name = [item[0].replace(' (dir.)', '') for item in directoractor_list]
stars_list = [stars.find('strong').text for stars in table.find_all('td', attrs = {'class': 'ratingColumn imdbRating'})]
data = list(zip(title, initial_release, director_name, stars_list))
df = pd.DataFrame(np.array(data), columns=headers)
````

#### Find the book name, price and stock availability as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'
````

````python
websites = []
for i in range(1, 51, 1):
    page = 'page-'+ str(i) + '.html'
    website = 'http://books.toscrape.com/catalogue/' + page
    websites.append(website)
````

````python
# your code here
df = pd.DataFrame()

for url in websites:
    html = requests.get(url).content
    soup = BeautifulSoup(html)
    table = soup.find_all('li', attrs = {'class' : 'col-xs-6'})
    
    book_name = [name.find_all('a')[0].find('img')['alt'] for name in table]
    price = [price.text for price in soup.find_all('p', attrs = {'class': 'price_color'})]
    stock = [stock.text.strip() for stock in soup.find_all('p', attrs = {'class' : 'instock availability'})]
    
    mini_df = pd.DataFrame({'book_name' : book_name,
                           'price' : price,
                           'stock' : stock})
    df = pd.concat([df, mini_df])
    df = df.reset_index(drop=True)
````
