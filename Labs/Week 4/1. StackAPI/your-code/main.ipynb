#### Question 1: Find the questions and answers of last month.

```python
import time
from datetime import datetime
d = datetime(2020, 3, 1, 0, 0)

unixtime = time.mktime(d.timetuple())
unixtime
```

```python
# your code here
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate =1580526000, todate = 1583031600)
questions
```

```python
SITE = StackAPI('stackoverflow')
answers = SITE.fetch('answers', fromdate =1580526000, todate = 1583031600)
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
d = datetime(2020, 3, 18, 0, 0)

unixtime = time.mktime(d.timetuple())
unixtime
```

```python
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate = 1584500400, min = '5', sort = 'votes' ,tagged = 'python')
questions
```
