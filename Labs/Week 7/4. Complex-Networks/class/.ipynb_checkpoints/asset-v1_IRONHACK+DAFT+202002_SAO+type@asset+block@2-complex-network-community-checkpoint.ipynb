# Finding Communities in Graphs

```python
import matplotlib.pyplot as plt
import networkx as nx
import warnings
warnings.filterwarnings("ignore")
plt.rcParams["figure.figsize"] = (7,7)
```

## Creating a sample graph (from the networkx database)

```python
G = nx.karate_club_graph()
pos = nx.spring_layout(G)
```

```python
nx.draw(G, pos=pos, with_labels=True)
plt.show()
```

```python
communities = nx.community.girvan_newman(G)
```

```python
next(communities)
```

```python
next(communities)
```

```python
next(communities)
```

```python
community = next(communities)
community
```

```python
communities = nx.community.girvan_newman(G)
community = next(communities)
community
```

```python
node_color = dict()

for node in community[0]:
    node_color[node] = 'r'

for node in community[1]:
    node_color[node] = 'b'

node_color = [node_color[node] for node in sorted(node_color)]
node_color
```

```python
nx.draw(G, pos=pos, node_color=node_color, with_labels=True)
plt.show()
```

```python
next(communities)
community = next(communities)
community
```

```python
node_color = dict()

for node in community[0]:
    node_color[node] = 'r'

for node in community[1]:
    node_color[node] = 'b'

for node in community[2]:
    node_color[node] = 'g'

for node in community[3]:
    node_color[node] = 'c'


node_color = [node_color[node] for node in sorted(node_color)]
node_color
```

```python
nx.draw(G, pos=pos, node_color=node_color, with_labels=True)
plt.show()
```
